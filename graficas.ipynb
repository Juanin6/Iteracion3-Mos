### Barras de costo, tiempo y memoria

```python
import pandas as pd
import matplotlib.pyplot as plt

# Construir DataFrame con los resultados proporcionados
data = {
    'size':      [15, 15, 19, 19, 24, 24],
    'method':    ['Pyomo', 'GA', 'Pyomo', 'GA', 'Pyomo', 'GA'],
    'objective': [17.16, 18.09, 21.74, 23.42, 25.24, 29.41],
    'time_s':    [41.78,  3.55, 1270.18, 4.47, 1269.22, 3.83]
}
df = pd.DataFrame(data)

# Ordenar tamaos de instancia
sizes = sorted(df['size'].unique())

# Separar mtodos
pyomo = df[df['method'] == 'Pyomo'].set_index('size').loc[sizes]
ga    = df[df['method'] == 'GA'].set_index('size').loc[sizes]

# Grfica de barras para Objetivo (Costo)
x = list(range(len(sizes)))
width = 0.35

plt.figure()
plt.bar([i - width/2 for i in x], pyomo['objective'], width)
plt.bar([i + width/2 for i in x], ga['objective'], width)
plt.xticks(x, sizes)
plt.xlabel('Nmero de clientes')
plt.ylabel('Objetivo (costo)')
plt.legend(['Pyomo', 'GA'])
plt.title('Comparacin de Objetivo por Instancia')
plt.tight_layout()
plt.show()

# Grfica de barras para Tiempo de cmputo
plt.figure()
plt.bar([i - width/2 for i in x], pyomo['time_s'], width)
plt.bar([i + width/2 for i in x], ga['time_s'], width)
plt.xticks(x, sizes)
plt.xlabel('Nmero de clientes')
plt.ylabel('Tiempo de cmputo (s)')
plt.legend(['Pyomo', 'GA'])
plt.title('Comparacin de Tiempo de Cmputo por Instancia')
plt.tight_layout()
plt.show()
```
